# LEO Environment Analysis
This notebook explores the historical evolution of objects in Low Earth Orbit (LEO), including active satellites, defunct satellites, and space debris. The insights gathered here will guide the development of AI-based collision detection and prevention strategies in the BumperSats project.


```python
import pandas as pd

df = pd.read_csv("/content/low-earth-orbits-objects.csv")
df.head()
# Rename columns for easier access
df.columns = ['entity', 'code', 'year', 'n_tracked_objects']

# Convert year to datetime (optional if we want time series plots)
df['year'] = pd.to_datetime(df['year'], format='%Y')
```

ðŸ“Š 1. Debris Growth Over Time
This visualization shows the exponential growth of tracked debris in LEO.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='year', y='n_tracked_objects', label='Debris Objects')
plt.title("Growth of Tracked Debris in Low Earth Orbit (LEO)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Number of Objects")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Insight: Debris Proliferation in LEO

Since the launch of Sputnik in 1957, the number of tracked objects in Low Earth Orbit (LEO) has exploded. By the early 1960s, fewer than 500 debris objects were tracked. Today, that number has grown into the tens of thousands, driven by:

- Collisions (e.g. Iridium 33 & Cosmos 2251)
- Anti-satellite weapon tests (e.g. China 2007, Russia 2021)
- Increasing launch activity from mega-constellations

This data reinforces the urgency of proactive collision avoidance systems like **BumperSats**.
